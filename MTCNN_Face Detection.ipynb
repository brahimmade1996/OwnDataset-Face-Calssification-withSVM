```python
# plot photo with detected faces using MTCNN
import cv2
import os
import numpy
from cv2 import imread
from cv2 import imshow
from cv2 import waitKey
from cv2 import destroyAllWindows
from cv2 import rectangle
from mtcnn.mtcnn import MTCNN
import sys

# load the MTCNN 3-STAGE NETWORK 
mtcnn = MTCNN()


def detect(pixels):
    # perform face detection using MTCNN
    bboxes = mtcnn.detect_faces(pixels)
    for box in bboxes:
        # extract
        x, y, width, height = box['box']
        x2, y2 = x + width, y + height
        # draw a rectangle over the pixels
        rectangle(pixels, (x, y), (x2, y2), (150,20,255), 1) #GBR
        #extract_face(x,y,x2,y2) #you can extract the faces also
        # show the image
    imshow('face detection',pixels)
    # keep the window open until we press a key
    waitKey(0)
    # close the window
    #destroyAllWindows()


# 1.load the photograph
#path = "./Will"
#for files in os.listdir(path):
#    print(files)
#    pixels = imread(path+"/"+files)
#    detect(pixels)
####################

# 2. Or enable live feed face detection
video_capture = cv2.VideoCapture(0)
while True:
    ret, frame = video_capture.read()
    colored = cv2.cvtColor(frame, cv2.COLOR_BGRA2BGR) #this converts the blue green red alpha channel (4 channels) to RGB (3 CHANNELS)
    #colored = cv2.resize(colored,(160,160)) #facenet expects images of size 160x160
    detect(colored)
    if cv2.waitKey(1) & 0xFF == ord('p') :
        break
video_capture.release() # When everything is done, release the capture
cv2.destroyAllWindows() #and destroy all windows

####################
```
